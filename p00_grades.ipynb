# Оценивание

Названия колонок в оценках за задачи:
* correct -- корректность решения задачи
* algo -- адекватность выбора алгоритма (насколько алгоритм подходящий)
* code -- качество кода (см. линтеры)
* doc -- наличие документации (комментарии и докстринги)
* intime -- факт сдачи работы вовремя

За каждый пункт выдаётся отметка от 0 до 1. Потом всё это когда-нибудь просуммируем и посмотрим, сколько кому каких баллов поставить в зачёт. Зачём будем выставлять по решению задач и по посещениям.

```python
# Всякий код

import numpy as np
import pandas as pd
from IPython.display import display_markdown

class ProblemGrades:
    """
    Класс для формирования табличек с оценками за задания.
    """
    columns = ['correct', 'algo', 'code', 'doc', 'intime']
    index = ['problem', 'nickname']
    # ['Корректность', 'Алгоритм', 'Код', 'Документация', 'Своевременность']
    
    def __init__(self):
        self.grades = pd.DataFrame(columns=self.columns + self.index)
        self.grades = self.grades.set_index(self.index)

    def assign(self, problem, nickname, **gradings):
        self.grades.loc[(problem, nickname), gradings.keys()] = gradings.values()

    def format_problem(self, problem):
        return self.grades.loc[problem]


pg = ProblemGrades()
```
